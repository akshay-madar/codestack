```python
## importing necessary packages 
from selenium.webdriver import Chrome
import pandas as pd

## Driver path near files directory
webdriver = "chromedriver.exe"

##invoking chrome driver
driver = Chrome(webdriver)
```

```python
url="https://www.hsx.com/security/feature.php?type=boxoffice" ## focussing and starting from the first page
driver.get(url) ## get the url
quotes = driver.find_element_by_xpath('//*[@id="container"]/div[2]/div[1]/div[2]/table/tbody') ## obtaining all quotes
Data_scraped=[]
for row_1 in quotes.find_elements_by_xpath(".//tr"):

    for td_prime in row_1.find_elements_by_xpath(".//td"):

        td_prime.find_element_by_xpath('//a[contains(@href, "security")]')
        Data_scraped.append(td_prime.text)
        #print(td_prime.text)
```

```python
Trips=[Data_scraped[i:i+8] for i in range(0,len(Data_scraped),8)]
Dataframe_scraped = pd.DataFrame(Trips,columns=['Sl_no','Movie Name','Symbol','Week Gross','Total Gross','Week','Price','Trade Button']) 
```
